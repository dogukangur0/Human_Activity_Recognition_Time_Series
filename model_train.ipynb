```python
import torch
import torch.nn as nn
from preprocessing import train_dataLoader
from cnn_lstm_model import CNN_LSTM_Model
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
device
```

```python
model = CNN_LSTM_Model(input_dim = 1152, hidden_dim = 128, output_dim = 64, num_classes = 6).to(device)
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr = 1e-3)
```

```python
EPOCHS = 3
for epoch in range(EPOCHS):
    train_loss, train_acc = 0, 0
    for X_batch, y_batch in train_dataLoader:
        X_batch, y_batch = X_batch.to(device), y_batch.to(device)
        output = model(X_batch)
        loss = loss_fn(output, y_batch)
        train_loss += loss.item()

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        output_label = torch.argmax(output, 1)
        train_acc += (output_label == y_batch).sum().item() / len(output_label)
        
    train_loss = train_loss / len(train_dataLoader)
    train_acc = train_acc / len(train_dataLoader)

    print(f"Epoch: {epoch}/{EPOCHS} | Train Loss: {train_loss} | Train Acc: {train_acc}")
```

```python
torch.save(model.state_dict(), "cnn_lstm_model.pth")
```
