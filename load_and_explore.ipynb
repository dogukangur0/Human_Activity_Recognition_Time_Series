```python
import numpy as np
import os
```

```python
def load_inertial_signals(folder_path, subset):
    inertial_signals = [
        "body_acc_x",
        "body_acc_y",
        "body_acc_z",
        "body_gyro_x",
        "body_gyro_y",
        "body_gyro_z",
        "total_acc_x",
        "total_acc_y",
        "total_acc_z"
    ]
    signals_data = []
    for inertial_signal in inertial_signals:
        file_name = f"{inertial_signal}_{subset}.txt"
        file_path = os.path.join(folder_path, file_name)
        data = np.loadtxt(file_path)
        signals_data.append(data)

    stacked = np.stack(signals_data, axis = 0).transpose(1,2,0)
    return stacked
```

```python
X_train = load_inertial_signals("UCI HAR Dataset/train/Inertial Signals/", "train") 
X_test = load_inertial_signals("UCI HAR Dataset/test/Inertial Signals/", "test")    
```

```python
X_train.shape # (N,C,L) -> (batch_size, channels, seq_len)
```

```python
y_train = np.loadtxt("UCI HAR Dataset/train/y_train.txt").astype(int) - 1
y_test = np.loadtxt("UCI HAR Dataset/test/y_test.txt").astype(int) - 1
```

```python
np.save("X_train.npy", X_train)
np.save("X_test.npy", X_test)
np.save("y_train.npy", y_train)
np.save("y_test.npy", y_test)
```
